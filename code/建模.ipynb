```python
def min_max_scaling(data, feature_range=(0, 1)):
    """
    对每列数据进行最大最小值缩放
    
    参数：
    - data：要进行缩放的数据，可以是 NumPy 数组、Pandas DataFrame 或其他类似数据结构
    - feature_range：缩放后的特征范围，默认为 (0, 1)
    
    返回值：
    - scaled_data：缩放后的数据
    """
    min_val = np.min(data, axis=0)
    max_val = np.max(data, axis=0)
    range_min, range_max = feature_range
    
    scaled_data = (data - min_val) / (max_val - min_val) * (range_max - range_min) + range_min
    
    return scaled_data
```

## 构建训练函数

```python
def feature_transformer(x,y,test_size=0.3,random_state=12):
    """
    负责分割数据集并转换onehot特征，返回转换后的稀疏矩阵和特征名
    """
    #1.重新命名列名
    #原列名
    cols=[ '面积(平方米)', '服务费', 'A地铁线', 'B地铁线', 'C地铁线', '大区标签', '品牌标签',
       '租赁方式标签', '装修标签', '朝向标签', '电梯_标签', '车位_标签', '用水_标签', '用电_标签', '燃气_标签',
       '采暖_标签', 'A付款方式标签', 'B付款方式标签', 'C付款方式标签', '楼层标签', '卧室', '客厅', '卫生间',
       '房+卫+厅', '房/总', '卫/总', '厅/总', '卧室面积', '户型', '是否地铁', '平均值特征1',
       '大区+街道平均值特征', '朝向标签平均值特征', 'A地铁线平均值特征', '大区英文平均值特征', '房屋配置聚类平均值特征2',
       '房屋装修聚类平均值特征3', '房屋费用聚类平均值特征4', '房屋便利聚类平均值特征5', '小区_标签']
    #新列名
    new_cols=[chr(65+s)+str(i) for s in range(len(cols)//10+1) for i in range(10)]
    new_cols=new_cols[:len(cols)]
    #特征名映射字典
    cols_map={k:v for k,v in zip(cols,new_cols)}
    #重新命名列
    x.columns=[cols_map[k] for k in x.columns]
    
    #2.分割数据集
    train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=test_size,random_state=random_state)
    
    #3.转换onehot特征
    #返回稀疏矩阵，有两个优点：
    #1.占用内存大幅减小，让并行成为可能，不然并行的话，内存爆掉
    #2.可以加速xgboost训练
    vector=DictVectorizer(sparse=True)
    x_train=vector.fit_transform(train_x.to_dict(orient='records'))
    x_test=vector.transform(test_x.to_dict(orient='records'))
    features=vector.get_feature_names()
    
    #4.构建原始特征对应的新下标字典  
    #原始特征名
    
    #离散列特征下标字典
    feature_map={k:[] for k in cols}
    for i in range(len(features)):
        for col in cols:
            if features[i].startswith(cols_map[col]):#如果新特征以老特征开头
                feature_map[col].append(i)
                break
    return (x_train,x_test,train_y,test_y),feature_map,cols_map
```

```python
from sklearn.metrics import mean_squared_error, r2_score

def train(cols, data, feature_map, num_round=500):
    """
    负责完成一次xgboost训练，返回测试集 RMSE, R²
    """
    # 1.获取数据集
    train_x, test_x, train_y, test_y = data
    
    # 获取原始特征对应的新特征下标
    index = []
    for col in cols:
        index.extend(feature_map[col])
    x_train = train_x[:, index]
    x_test = test_x[:, index]
    
    # 3.构建数据格式
    # 构建DMatrix数据，可以有效利用硬盘缓存，减少内存占用
    dtrain = xgb.DMatrix(x_train, train_y)
    dtest = xgb.DMatrix(x_test, test_y)
    
    # 4.设置训练参数
    param = {
        'max_depth': 5, 
        'eta': 0.01, 
        'verbosity': 1, 
        'objective': 'reg:linear',
        'silent': 1,
        'gamma': 0.01,
        'min_child_weight': 1
    }
    
    # 5.模型训练
    bst = xgb.train(param, dtrain, num_boost_round=num_round)
    
    # 6.模型预测
    preds = bst.predict(dtest)
    
    # 7.模型评估
    rmse = np.sqrt(mean_squared_error(test_y, preds))
    r2 = r2_score(test_y, preds)
    
    return rmse, r2
```

```python
import numpy as np

# 构造筛选特征函数
def select_features(cols, min_rmse, min_r2):
    include_features = cols  # 初始时包含所有特征
    exclude_features = []  # 初始时排除特征为空
    best_rmse = float('inf')  # 初始化最佳 RMSE 为正无穷大
    best_r2 = float('-inf')  # 初始化最佳 R² 为负无穷大
    
    for i in range(len(cols)):
        # 选中的特征
        features = list(set(cols) - set([cols[i]]))  # 移除当前特征
        print("开始第{}次训练:".format(i))
        print("未选中特征：", cols[i])
        rmse, r2 = train(features, d, f_map, 2500)  # 训练模型并返回评估指标
        print("开始第{}次训练测试集成绩 -  RMSE: {:.4f}, R²: {:.4f}".format(i, rmse, r2))
        print("-" * 10)
        
        # 只有当 RMSE 较低且 R² 较高时才更新最佳特征
        if rmse <= best_rmse and r2 >= best_r2:
            exclude_features = cols[i]  # 更新排除特征列表
            include_features = features  # 更新包含特征列表
            best_rmse = rmse  # 更新最佳评分为当前 RMSE
            best_r2 = r2  # 更新最佳 R² 为当前 R²
        
        # 如果剩下最后一个特征，则跳出循环
        if len(features) == 1:
            break
    
    return best_rmse, best_r2, include_features, exclude_features  # 返回最佳评分，选中特征列表和排除特征列表
```

```python
from sklearn.metrics import mean_squared_error, r2_score

def train_cv(data, target, params, num_round=200, k_fold=5, silent=0):
    """
    负责完成一种参数组合的情况下 k_flod 折交叉验证的平均 RMSE 值和 R2 值
    """
    rmses = []
    r2s = []
    
    # 数据分割
    kfold = KFold(n_splits=k_fold, random_state=None, shuffle=True)
    for i, (train_index, val_index) in enumerate(kfold.split(data, target)):
        train_x, val_x, train_y, val_y = data[train_index, :], data[val_index, :], target[train_index], target[val_index]
        # 构建 DMatrix 数据
        dtrain = xgb.DMatrix(train_x, train_y)
        dtest = xgb.DMatrix(val_x, val_y)
        
        if silent == 0:
            print("开始第{}/{}折验证：".format(i, k_fold))
        
        # 模型训练
        bst = xgb.train(params, dtrain, num_round)
        
        # 模型预测
        preds = bst.predict(dtest)
        preds = preds  # 这里不进行转换，保留原始预测值
        y_true = val_y
        
        # 计算 RMSE
        rmse = np.sqrt(mean_squared_error(y_true, preds))
        if silent == 0:
            print("第{}/{}折验证 RMSE：{}".format(i, k_fold, rmse))
        
        # 计算 R2
        r2 = r2_score(y_true, preds)
        if silent == 0:
            print("第{}/{}折验证 R2：{}".format(i, k_fold, r2))
        
        rmses.append(rmse)
        r2s.append(r2)
    
    avg_rmse = sum(rmses) / k_fold
    avg_r2 = sum(r2s) / k_fold
    
    return avg_rmse, avg_r2
```

```python
def search_params(x, y, params_grid, n_estimators=200, cv=3, silent=0):
    min_rmse = 9999
    best_params = None
    best_r2 = -9999
    
    params_list = [[]]
    
    # 根据参数表格构建所有参数组合
    for k, v in params_grid.items():
        if isinstance(v, list):
            temp = params_list
            params_list = [i+[j] for j in v for i in temp]
        else:
            params_list = [i+[v] for i in params_list]
    params_list = [{k: v for k, v in zip(params_grid.keys(), v_list)} for v_list in params_list]
    
    for i, params in zip(range(len(params_list)), params_list):
        if silent == 0:
            print("开始实验第{}组参数：".format(i), params)
        rmse, r2 = train_cv(data=x, target=y, params=params, num_round=n_estimators, k_fold=cv)
        if silent == 0:
            print("第{}组参数平均 RMSE：{}".format(i, rmse))
            print("第{}组参数平均 R2：{}".format(i, r2))
            print("-" * 50)
        if rmse < min_rmse:
            min_rmse = rmse
            best_params = params
            best_r2 = r2  # 更新最好的 R2
        elif rmse == min_rmse and r2 > best_r2:
            best_params = params
            best_r2 = r2  # 更新最好的 R2
    return min_rmse, best_params, best_r2
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer, mean_squared_error, r2_score

def search_params_suiji_xgboost(x, y, params_grid, n_estimators=200, cv=3, silent=0):
    # 定义评估指标
    scoring = {'RMSE': make_scorer(mean_squared_error, squared=False),
               'R2': make_scorer(r2_score)}
    
    # 创建 XGBoost 回归器模型
    xgb_model = XGBRegressor(n_estimators=n_estimators)

    # 定义随机搜索对象
    random_search = RandomizedSearchCV(estimator=xgb_model, param_distributions=params_grid, 
                                       n_iter=10, scoring=scoring, cv=cv, 
                                       refit='RMSE', random_state=42, 
                                       verbose=2, n_jobs=-1)
    
    # 执行随机搜索
    random_search.fit(x, y)
    
    # 输出最佳参数和评估指标
    print("Best Parameters:", random_search.best_params_)
    print("Best RMSE:", random_search.best_score_)
    print("Best R2:", random_search.cv_results_['mean_test_R2'][random_search.best_index_])
    
    return random_search.best_score_, random_search.best_params_, random_search.cv_results_['mean_test_R2'][random_search.best_index_]
```

```python
from skopt import BayesSearchCV
from skopt.space import Real, Integer, Categorical

def search_params_bayesian(x, y, params_grid, n_estimators=200, cv=3, silent=0):
    # 定义搜索空间
    search_space = {
    'objective': Categorical(params_grid['objective']),
    'eta': Real(0.01, 0.5, prior='log-uniform'),
    'gamma': Real(0.01, 0.1, prior='uniform'),
    'silent': Categorical(params_grid['silent']),
    'max_depth': Integer(15, 30, prior='uniform'),
    'min_child_weight': Integer(0.5, 5, prior='uniform'),
    'subsample': Real(0.6, 0.8, prior='uniform'),
    'colsample_bytree': Real(0.6, 0.8, prior='uniform')
}

    
    # 创建 XGBRegressor 模型
    xgb_model = XGBRegressor(n_estimators=n_estimators)
    
    # 定义贝叶斯搜索对象
    bayes_search = BayesSearchCV(estimator=xgb_model, search_spaces=search_space, 
                                 n_iter=100, scoring='neg_root_mean_squared_error', 
                                 cv=cv, n_jobs=-1, verbose=2, random_state=42)
    
    # 执行贝叶斯搜索
    bayes_search.fit(x, y)
    
    # 输出最佳参数和评估指标
    print("Best Parameters:", bayes_search.best_params_)
    print("Best RMSE:", -bayes_search.best_score_)
    
    # 计算 R2 分数
    best_model = bayes_search.best_estimator_
    y_pred = best_model.predict(x)
    r2 = r2_score(y, y_pred)
    print("Best R2:", r2)
    
    return -bayes_search.best_score_, bayes_search.best_params_, r2
```
