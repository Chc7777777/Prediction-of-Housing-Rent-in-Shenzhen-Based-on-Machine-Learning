###  面积对租金的影响

```python
####面积对租金的影响
from pyecharts import options as opts
from pyecharts.charts import Scatter

def draw_area_rent_scatter(data, x_column, y_column, x_name, y_name):
    # 准备数据
    x_data = data[x_column].tolist()
    y_data = data[y_column].tolist()

    # 创建散点图实例
    scatter = (
        Scatter(init_opts=opts.InitOpts(theme=ThemeType.VINTAGE))
        .add_xaxis(xaxis_data=x_data)
        .add_yaxis(series_name=y_name, y_axis=y_data)
        .set_global_opts(
            title_opts=opts.TitleOpts(title="{}与{}关系散点图".format(x_name, y_name)),
            xaxis_opts=opts.AxisOpts(name=x_name),
            yaxis_opts=opts.AxisOpts(name=y_name),
            tooltip_opts=opts.TooltipOpts(trigger='axis', axis_pointer_type='cross'),
        )
    )

    return scatter

# 绘制面积与租金关系散点图
scatter = draw_area_rent_scatter(data_pd, '面积(平方米)', '租金', '面积(平方米)', '租金')
scatter.render("E:\\IT\\毕设\\photo\\area_rent_scatter.html")
```

```python
import pandas as pd
from pyecharts.charts import Bar
from pyecharts.globals import ThemeType

def draw_area_rent_bar(data, x_column, y_column, x_name, y_name):
    # 将面积范围分组，并计算每个范围内的租金均值或中位数
    bins = [0, 20, 50, 80, 100, 150, 200, float('inf')]
    labels = ['0-20', '20-50', '50-80', '80-100', '100-150', '150-200', '200以上']
    data['面积区间'] = pd.cut(data[x_column], bins=bins, labels=labels)
    grouped_data = data.groupby('面积区间')[y_column].mean().reset_index()

    # 提取分组后的面积范围和对应的租金均值
    x_data = grouped_data['面积区间'].tolist()
    y_data = grouped_data[y_column].tolist()

    # 创建柱形图实例
    bar = (
        Bar(init_opts=opts.InitOpts(theme=ThemeType.VINTAGE))
        .add_xaxis(xaxis_data=x_data)
        .add_yaxis(series_name=y_name, y_axis=y_data, label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="{}与{}关系柱形图".format(x_name, y_name)),
            xaxis_opts=opts.AxisOpts(name=x_name),
            yaxis_opts=opts.AxisOpts(name=y_name),
        )
    )
    return bar

# 绘制面积与租金关系柱形图
bar = draw_area_rent_bar(data_pd, '面积(平方米)', '租金', '面积(平方米)', '租金')
bar.render("E:\\IT\\毕设\\photo\\\\area_rent_bar.html")
```

```python
import pandas as pd
from pyecharts.charts import Bar
from pyecharts.globals import ThemeType
from pyecharts import options as opts

def draw_area_district_bar(data, area_column, district_column, rent_column, x_name, y_name):
    # 划分面积区间
    bins = [0, 20, 50, 80, 100, 150, 200, float('inf')]
    labels = ['0-20', '20-50', '50-80', '80-100', '100-150', '150-200', '200以上']
    data['面积区间'] = pd.cut(data[area_column], bins=bins, labels=labels)

    # 计算每个面积区间和大区的租金均值
    grouped_data = data.groupby(['面积区间', district_column])[rent_column].mean().reset_index()

    # 获取所有大区编号
    districts = sorted(data[district_column].unique())

    # 绘制柱形图
    bar = (
        Bar(init_opts=opts.InitOpts(theme=ThemeType.VINTAGE))
        .add_xaxis(xaxis_data=labels)
        .set_global_opts(
            title_opts=opts.TitleOpts(title="{}与{}关系柱形图".format(x_name, y_name)),
            xaxis_opts=opts.AxisOpts(
                type_="category",
                axislabel_opts=opts.LabelOpts(rotate=30, interval=0),
                name=x_name,
            ),
            yaxis_opts=opts.AxisOpts(name=y_name),
            datazoom_opts=opts.DataZoomOpts(
                is_show=True,
                type_="slider",
                range_start=0,
                range_end=100,
            ),
        )
    )

    # 添加每个大区的租金数据
    for district in districts:
        district_data = grouped_data[grouped_data[district_column] == district]
        bar.add_yaxis(
            series_name="大区{}".format(district),
            y_axis=district_data[rent_column].tolist(),
            label_opts=opts.LabelOpts(is_show=True, position="top", formatter="{c}"),
            stack="stack",
        )

    return bar

# 绘制面积和大区对租金的影响柱形图
bar = draw_area_district_bar(data_pd, '面积(平方米)', '大区标签', '租金', '面积区间', '租金')
bar.render("E:\IT\毕设\photo\双因素对租金的影响\\area_district_rent_bar.html")
```
